```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate
```

```python
pi = -np.array([4.3,4.4,4.2,3.2,2.6,2.2,4.1,3.0,4.2,5.3,5.4,4.1,3.6,2.8,3.4])
s2 = -np.array([5.1,5.0,4.5,4.1,3.1,3.7,4.2,4.8,5.0,6.2,6.4,5.4,5.0,4.7,4.9])
xpos = 10*np.arange(len(pi))
```

```python
save = True
xnew = np.linspace(xpos[0],xpos[-1],100)
spl = interpolate.interp1d(xpos, pi, kind='cubic')
pi_smooth = spl(xnew)
spl = interpolate.interp1d(xpos, s2, kind='cubic')
s2_smooth = spl(xnew)

fig, ax1 = plt.subplots(figsize=(10,4))

ax1.set_xlabel('chr8 position (Mb)')
ax1.set_ylabel('polygenicity', color="C0")
ax1.plot(xnew,pi_smooth,color="C0")
ax1.tick_params(axis='y', labelcolor="C0")
ax1.set_ylim((-5,-2))


ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

ax2.set_ylabel('discoverability', color="C1")
ax2.plot(xnew, s2_smooth, color="C1")
ax2.tick_params(axis='y', labelcolor="C1")
ax2.set_ylim((-6,-3))

fig.tight_layout()
if save:
    plt.savefig("regional_cmma_concept.svg")
    plt.savefig("regional_cmma_concept.png")
plt.show()
```
