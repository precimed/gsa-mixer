```python
import numpy as np
import os
import pandas as pd
from collections import defaultdict
```

```python
res_dir1 = "/mnt/c/Users/Алексей/OneDrive/projects/uio/cmma/tests/experiment_8"
res_dir2 = "/mnt/c/Users/Алексей/OneDrive/projects/uio/cmma/tests/experiment_9"
res_dir3 = "/mnt/c/Users/Алексей/OneDrive/projects/uio/cmma/tests/experiment_9"
res_dir4 = "/mnt/c/Users/Алексей/OneDrive/projects/uio/cmma/tests/experiment_9"
trait = "pgc_scz_2014_eur"
npz_file_prefix1 = f"optimize.1000genomes503eur9m.annot.utr5utr3exon_other.sumstats.{trait}.experiment_8.rsubset_500000.no_mhc"
npz_file_prefix2 = f"optimize.1000genomes503eur9m.annot.utr5utr3exon_other.sumstats.{trait}.experiment_9.same_sigma"
npz_file_prefix3 = f"optimize.1000genomes503eur9m.annot.utr5utr3exon_other.sumstats.{trait}.experiment_9.same_p"
npz_file_prefix4 = f"optimize.1000genomes503eur9m.annot.utr5utr3exon_other.sumstats.{trait}.experiment_9.same_p_and_sigma"
n_runs = 15
npz_files1= [os.path.join(res_dir1, npz_file_prefix1 + f".{i}" + ".npz") for i in range(1, n_runs+1)]
npz_files2 = [os.path.join(res_dir2, npz_file_prefix2 + f".{i}" + ".npz") for i in range(1, n_runs+1)]
npz_files3 = [os.path.join(res_dir3, npz_file_prefix3 + f".{i}" + ".npz") for i in range(1, n_runs+1)]
npz_files4 = [os.path.join(res_dir4, npz_file_prefix4 + f".{i}" + ".npz") for i in range(1, n_runs+1)]
group_names = ["all_free", "same_sigma", "same_p", "same_p_and_sigma"]
group_n_pars = [5,4,4,3]
n_samples = 500000./2
ic = "aic" # "aic"
```

```python
res_dict = defaultdict(list)
n_run = 1
for npz_files in zip(npz_files1, npz_files2, npz_files3, npz_files4):
    res_dict["run"].append(n_run)
    n_run += 1
    for i, npz_f in enumerate(npz_files):
        res = np.load(npz_f)
        fun = res.get("fun").item(0)
        if ic == "bic":
            ic_val = np.log(n_samples)*group_n_pars[i] + 2*n_samples*fun
        elif ic == "aic":
            ic_val = 2*group_n_pars[i] + 2*n_samples*fun
        res_dict[group_names[i]].append(ic_val)
```

```python
df = pd.DataFrame(res_dict)
df["all_free_VS_same_sigma"] = df["all_free"] - df["same_sigma"]
df["all_free_VS_same_p"] = df["all_free"] - df["same_p"]
df["all_free_VS_same_p_and_sigma"] = df["all_free"] - df["same_p_and_sigma"]
df
```
